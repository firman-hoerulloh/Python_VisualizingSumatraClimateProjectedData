# Visualisasi Data

### Grafik Precipitation Tahun 2030 Per Bulan

```python
# Resample data ke frekuensi bulanan dan ambil rata-rata untuk setiap bulan
df_monthly = df_2030['precipitation'].resample('M').mean()

# Visualisasi time series plot
plt.figure(figsize=(10, 6))

# Plot 'precipitation'
plt.plot(df_monthly.index, df_monthly, marker='o', linestyle='-')

# Atur label sumbu dan judul
plt.xlabel('Month')
plt.ylabel('Precipitation')
plt.title('Monthly Precipitation in Sumatra (2030)')

# Tampilkan plot
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Time Series Plot: yang menunjukkan bagaimana 'precipitation' berubah dari bulan ke bulan sepanjang tahun 2030

Keterangan:
Precipitation merupakan istilah yang digunakan dalam ilmu atmosfer untuk menggambarkan segala bentuk air yang jatuh dari atmosfer ke permukaan bumi. Ini termasuk hujan, salju, hujan es, embun beku, atau bahkan kondensasi langsung dari uap air di udara menjadi titik-titik air atau kristal salju.

### Histogram Data Sepanjang Tahun 2030

```python
# Buat histogram
plt.figure(figsize=(12, 8))

# Histogram untuk 'precipitation'
plt.subplot(3, 2, 1)
plt.hist(df_2030['precipitation'], bins=30, color='blue', alpha=0.7)
plt.title('Precipitation Distribution (2030)')
plt.xlabel('Precipitation')
plt.ylabel('Frequency')

# Tampilkan plot
plt.show()
```

Histogram pada gambar tersebut menunjukkan distribusi frekuensi curah hujan di tahun 2030. Sumbu x pada histogram menunjukkan jumlah curah hujan, dalam milimeter, dan sumbu y menunjukkan frekuensi dari tiap jumlah curah hujan tersebut.

Histogram menunjukkan bahwa sebagian besar curah hujan di tahun 2030 berada di antara 0,006 dan 0,008 milimeter. Ada juga curah hujan yang signifikan antara 0,008 dan 0,010 milimeter. Namun, curah hujan di atas 0,010 milimeter sangat sedikit.

```python
# Buat histogram
plt.figure(figsize=(12, 8))

# Histogram untuk 'temperature'
plt.subplot(3, 2, 2)
plt.hist(df_2030['temperature'], bins=30, color='red', alpha=0.7)
plt.title('Temperature Distribution (2030)')
plt.xlabel('Temperature')
plt.ylabel('Frequency')

# Tampilkan plot
plt.show()
```

Histogram pada gambar tersebut menunjukkan bahwa suhu paling sering terjadi di tahun 2030 adalah 27 derajat Fahrenheit. Ada juga jumlah hari yang signifikan dengan suhu antara 26,6 dan 27,4 derajat Fahrenheit. Namun, sangat sedikit hari dengan suhu di bawah 26,4 derajat Fahrenheit atau di atas 27,6 derajat Fahrenheit.

```python
# Buat histogram
plt.figure(figsize=(12, 8))

# Histogram untuk 'max temperature'
plt.subplot(3, 2, 3)
plt.hist(df_2030['max temperature'], bins=30, color='green', alpha=0.7)
plt.title('Max Temperature Distribution (2030)')
plt.xlabel('Max Temperature')
plt.ylabel('Frequency')

# Tampilkan plot
plt.show()
```

Histogram tersebut menunjukkan bahwa suhu maksimum yang paling sering terjadi pada tahun 2030 adalah sekitar 30.4 derajat. Suhu maksimum yang lebih rendah dan lebih tinggi dari ini tampaknya kurang umum. Ini memberikan gambaran tentang bagaimana suhu maksimum didistribusikan selama tahun 2030.

```python
# Buat histogram
plt.figure(figsize=(12, 8))

# Histogram untuk 'wind speed'
plt.subplot(3, 2, 4)
plt.hist(df_2030['wind speed'], bins=30, color='orange', alpha=0.7)
plt.title('Wind Speed Distribution (2030)')
plt.xlabel('Wind Speed')
plt.ylabel('Frequency')

# Tampilkan plot
plt.show()
```

Dari histogram ini, kita dapat menyimpulkan bahwa kecepatan angin tertentu yang paling sering terjadi pada tahun 2030 berada dalam rentang 2.50 sampai 3.00. Kecepatan angin yang lebih rendah dan lebih tinggi dari ini tampaknya kurang umum. Ini memberikan gambaran tentang bagaimana kecepatan angin didistribusikan selama tahun 2030.

```python
# Buat histogram
plt.figure(figsize=(12, 8))

# Histogram untuk 'relative humidity'
plt.subplot(3, 2, 5)
plt.hist(df_2030['relative humudity'], bins=30, color='purple', alpha=0.7)
plt.title('Relative Humidity Distribution (2030)')
plt.xlabel('Relative Humidity')
plt.ylabel('Frequency')

# Tampilkan plot
plt.show()
```

Histogram tersebut menunjukkan distribusi kelembapan relatif di sebuah ruangan. Sumbu x merepresentasikan kelembapan relatif, dan sumbu y merepresentasikan frekuensi tiap kelembapan relatif tersebut.

Histogram menunjukkan bahwa kelembapan relatif yang paling umum berada di antara 81.5% dan 82%. Tingkat kelembapan relatif ini dianggap ideal untuk kenyamanan manusia.

Histogram juga menunjukkan adanya kelembapan relatif yang jauh lebih tinggi atau lebih rendah dari rata-rata. Kelembapan relatif ini meski kurang umum, tetap dapat mempengaruhi kenyamanan manusia. Misalnya, kelembapan relatif yang sangat tinggi dapat membuat pernapasan terganggu, sementara kelembapan relatif yang sangat rendah dapat membuat kulit dan selaput lendir kering.

### Perbandingan Temperature dengan Precipitation

```python
# Buat scatter plot untuk membandingkan 'temperature' dengan 'precipitation'
plt.figure(figsize=(10, 6))
plt.scatter(df['temperature'], df['precipitation'], color='blue', alpha=0.5)

# Atur label sumbu dan judul
plt.xlabel('Temperature')
plt.ylabel('Precipitation')
plt.title('Scatter Plot: Temperature vs Precipitation')

# Tampilkan plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

Dari Grafik tersebut tidak terlihat adanya pola yang jelas antara kedua variabel tersebut.  Beberapa titik dengan suhu tinggi memiliki curah hujan tinggi, sementara yang lainnya tidak.  Demikian pula, beberapa titik dengan suhu rendah memiliki curah hujan tinggi, sementara yang lainnya tidak.

### Perbandingan Precipitation dengan Relative Humidity

```python
# Buat scatter plot untuk membandingkan 'precipitation' dengan 'relative humidity'
plt.figure(figsize=(10, 6))
plt.scatter(df['precipitation'], df['relative humudity'], color='blue', alpha=0.5)

# Atur label sumbu dan judul
plt.xlabel('Precipitation')
plt.ylabel('Relative Humidity')
plt.title('Scatter Plot: Precipitation vs Relative Humidity')

# Tampilkan plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

Dari Grafik tersebut dapat dilihat titik-titik lebih terkonsentrasi dalam pola diagonal dari sudut kiri bawah grafik menuju sudut kanan atas. Ini menunjukkan adanya korelasi positif antara curah hujan dan kelembaban relatif.

### Heatmap Korelasi Data Keseluruhan

```python
# Buat heatmap untuk korelasi antara variabel-variabel utama
plt.figure(figsize=(10, 8))
sns.heatmap(df[['precipitation', 'temperature', 'max temperature', 'wind speed', 'relative humudity']].corr(), annot=True, cmap='coolwarm', fmt=".2f")

# Atur judul
plt.title('Heatmap Data Proyeksi Iklim Sumatera')

# Tampilkan plot
plt.tight_layout()
plt.show()
```

### Grafik Rata-Rata Precipitation Per Tahun

```python
# Resample data ke frekuensi tahunan dan ambil rata-rata untuk setiap tahun
yearly_precipitation = df['precipitation'].resample('Y').mean()

# Visualisasi rata-rata curah hujan per tahun
plt.figure(figsize=(10, 6))
plt.plot(yearly_precipitation.index.year, yearly_precipitation, marker='o', linestyle='-')

# Atur label sumbu dan judul
plt.xlabel('Year')
plt.ylabel('Average Precipitation')
plt.title('Average Precipitation Per Year')

# Tampilkan plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Grafik Rata-Rata Kecepatan Angin Per Tahun

```python
# Resample data ke frekuensi tahunan dan ambil rata-rata untuk setiap tahun
yearly_wind_speed = df['wind speed'].resample('Y').mean()

# Visualisasi rata-rata kecepatan angin per tahun dengan bar plot
plt.figure(figsize=(10, 6))
plt.bar(yearly_wind_speed.index.year, yearly_wind_speed)

# Atur label sumbu dan judul
plt.xlabel('Year')
plt.ylabel('Average Wind Speed')
plt.title('Average Wind Speed Per Year (Bar Plot)')

# Tampilkan plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Perbandingan Temperature dan Precipitation

```python
# Mengelompokkan data berdasarkan tahun dan menghitung rata-rata tahunan
df['tahun'] = df.index.year
df_yearly_mean = df.groupby('tahun').mean().reset_index()

# Memplot diagram garis untuk rata-rata tahunan temperature dan precipitation
plt.figure(figsize=(10, 6))

plt.plot(df_yearly_mean['tahun'], df_yearly_mean['max temperature'], color='blue', label='max temperature')
plt.plot(df_yearly_mean['tahun'], df_yearly_mean['temperature'], color='green', label='temperature')

plt.title('Yearly Mean Temperature vs Precipitation')
plt.xlabel('Year')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)
plt.show()
```
